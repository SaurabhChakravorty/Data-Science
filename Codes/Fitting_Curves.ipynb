## Fitting Curves

#### Generating points with a polynomial degree

##### Importing all important libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
def polynomial_features_simple(m,n):
    """
    This function takes "m" data points and generates polynomial output of degree "n" with a plot of function 
    
    """
    # Generating m data points with in the value "0" to "n"
    X = np.linspace(0,100,m).tolist()
    # For storing the values intialised aand generated
    y = []
    
    # Loop to get values
    y = [np.power(i,n) for i in X]
    
    # Generating plot from the figure
    fig = plt.figure(figsize=(7,7))
    #plt.yticks(range(int(np.min(y)), int(np.max(y))))
    plt.plot(X,y,label ='Polynomial with degree '+str(n))
    plt.legend()
    plt.show()
    
   # return the values 
    return X,y
```

```python
# Let's get the input and output values
i,o = polynomial_features_simple(100,10)
```

#### Generating random samples of polynomial degree n and (n-1) points for plotting

```python
def polynomial_features_degrees(n):
    """
    This function maps and plot n-1 degree points for polynomial points and performs a linear regression 
    
    """
    X = np.linspace(0,100,n-1)

    # Generating polynomial variables for output y
    y =  X*6 +  np.random.randn(n-1) * 100 + 100   # Can use any function use random x for convenience

    
    # Generating polynomial variable of 'n' features
    for i in range(1,n+1):
        # Pipeline for fitting with degree i
        poly_features = PolynomialFeatures(degree = i)  # Polynomial feature
        lm = LinearRegression()                         # Regression line plot
        # Fitting for model with polynomial of degree i
        X_poly = poly_features.fit_transform(X.reshape(-1,1))
        # Applying regression on model
        lm.fit(X_poly,y.reshape(-1, 1))
        
        # Predicitng values after fitting
        model_pred = lm.predict(X_poly)
        
        # Now plotting the graph
        plt.figure(figsize=(10,8))
        plt.scatter(X, y)
        plt.plot(X, model_pred,label ='Polynomial with degree '+str(i))
        plt.legend()
        plt.show()
        print(r2_score(y, model_pred))
        print("The R2 score of the model is %0.2f " %(r2_score(y, model_pred)))
```

Just add how much polynomial degrees required(n) the points will always be n-1

```python
polynomial_features_degrees(12)
```
